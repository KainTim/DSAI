## Colors and Colormaps

## Farben



(from https://www.virtualartacademy.com/three-components-of-color/)

### Farbton

* Ist der *Basiston* der Farbe (zBsp.: Gelb, Rot, Blau, Grün, Violett, etc.)
* Wird **nur** durch das RGB Verhältnis bestimmt
* Farbton bleibt gleich, auch wenn sich die Intensität oder die Helligkeit ändert 

### Sättigung

* Gibt an, wie *grau* eine Farbe ist
* Je weiter entfernt eine Farbe von der Gleichverteilung ist, desto gesättigter ist sie
* Eine maximal gesättigte Farbe ist zum Beispiel Rot (255, 0, 0), eine niedrig gesättigte Farbe ist (100, 100, 100).

### Helligkeit

* Gibt an, wie hell/dunkel die Farbe ist

Basierend auf https://manifold-net.translate.goog/doc/mfd8/colors_as_hue_saturation_and_brightness.htm?_x_tr_sl=en&_x_tr_tl=de&_x_tr_hl=de&_x_tr_pto=rq können wir auch folgende Formeln für die Sättigung und Helligkeit verwenden. Dabei verwenden wir $M=\max\{R,G,B\}$ und $m=\min\{R,G,B\}$.

$$L = \frac{M+m}{2}$$

$$S = \begin{cases}
    0& R=G=B\\
    255 \cdot \frac{M-m}{M+m}& L<128\\
    255 \cdot \frac{M-m}{511 - (M+m)}
\end{cases}$$

wobei $L$ die Helligkeit und $S$ die Sättigung beschreibt.

## Colormaps

* Es gibt mehrere Arten von Colormaps:
    * Sequential
    * Diverging
    * Categorical



(from https://medium.com/@oksana.zamyslova/data-visualisation-3d430868bfa5)

### Sequential Scale

* **Einsatz**: Darstellung quantitativer Zusammenhänge
    * **Beispiele:** Epidemie- oder Pandemiezahlen (Corona) bzw. Heatmaps, Zeitreihen (Durschnittstemperatur), Wahlergebnisse in einzelnen Gemeinden
* **Farbverlauf**: Farben wechseln von dunkel auf hell (oder umgekehrt) bzw. von geringer Intensität zu hoher Intensität (oder umgekehrt)
* **Vorteile**:
    * Einfache Interpretation
* **Nachteile**:
    * Barrierefreiheit: Schwer für blinde Menschen
    * Begrenzte Differenzierung: Feine Unterschiede können schwer/nicht encodiert werden
    * Verzerrung: Zu helle oder dunkle Bereiche könnten visuell überbetont werden, selbst wenn die Werte nur leicht variieren.

### Diverging Scale

* **Einsatz**: Visualisierung von Daten, die sich um einen zentralen, neutralen Wert gruppieren, i.e. Abweichungen von einem Referenzwert.
    * **Beispiele**: Temperaturabweichungen von einem Mittelwert, Finanzdaten (Gewinn/Verlust), politische Umfragen (z.B. Zustimmung/Ablehnung), Umweltdaten (z.B. pH-Werte)
* **Farbverlauf**: Zwei Farbtöne bewegen sich symmetrisch von einem neutralen Mittelpunkt in gegensätzliche Richtungen (z.B. Rot für negative und Blau für positive Werte), wobei die Intensität zunimmt.
* **Zentraler Referenzpunkt**: Die Skala betont eine Mitte (z.B. Nullpunkt oder Durchschnitt), von der positive und negative Werte abweichen.
* **Vorteile**:
    * Unterschiede von einem neutralen Punkt sind sofort erkennbar.
    * Neutralität: Positiv und negativ werden gleichwertig hervorgehoben, was eine neutrale Präsentation unterstützt.
* **Nachteile**:
    * Farbverwirrung: Ein unklarer zentraler Punkt oder Schlecht gewählte Farben kann zu Missverständnissen führen.
    * Barrierefreiheit: Bestimmte Farbenpaare (z.B. Rot-Grün) sind für farbenblinde Menschen schwer zu unterscheiden.
    * Begrenzte Differenzierung: Feine Unterschiede können schwer/nicht encodiert werden
* **Best Practices**:
    * Der zentrale Wert soll neutral sein, zBsp. weiß oder grau.
    * Richtige Farbwahl: Besser Rot-Blau als Rot-Grün (Farbblindheit)

### Categorical Scale

* **Einsatz**: Darstellung von Daten, die unabhängige Kategorien oder Gruppen repräsentieren, bei denen keine natürliche Reihenfolge besteht.
    * **Beispiele**: Umfragen mit mehreren Antwortmöglichkeiten
* **Farbverlauf**: Jede Kategorie wird durch eine klar voneinander abgegrenzte Farbe dargestellt. Die Farben sollten sich maximal voneinander unterscheiden.
* **Vorteile**:
    * Eindeutige Unterscheidung: Kontrastreicher Farben ermöglicht eine klare visuelle Trennung der Kategorien.
    * Ideal für qualitative Daten: Funktioniert gut, wenn es keine logische Reihenfolge oder Skala gibt.
    * Flexibel: Kann in Diagrammen, Karten, Balkendiagrammen und mehr verwendet werden, wo es wichtig ist, jede Kategorie deutlich zu erkennen.
* **Nachteile**:
    * Begrenzte Anzahl an Farben: Bei zu vielen Kategorien kann es schwierig werden, ausreichend unterscheidbare Farben zu finden, ohne dass die Farben sich zu ähnlich werden.
    * Barrierefreiheit: Farbpaletten müssen sorgfältig gewählt werden, da bestimmte Farben (z.B. Rot-Grün) für Menschen mit Farbsehschwäche schwer zu unterscheiden sind.
    * Keine Abbildung von Intensität: Im Gegensatz zu sequenziellen oder divergenten Skalen kann eine kategoriale Skala keine quantitativen Unterschiede darstellen.
* **Best Practices**:
    * Begrenzte Kategorien: Ideal für Datensätze mit einer moderaten Anzahl von Kategorien (in der Regel weniger als 10). Für größere Mengen sollten alternative Methoden (z.B. Texturen oder Symbole) in Kombination mit Farben eingesetzt werden.
    * Farbkontraste: Gut unterscheidbare Farben sollen verwendet werden. Tools wie [Colorbrewer](https://colorbrewer2.org/) können helfen
    * Konsistenz: Alle Visualisierungen sollten die gleichen Farben verwendet werden um Verwirrung in der Zuordnung zu vermeiden.
    * Unterstützende Labels: Farben alleine möglicherweise nicht ausreichend. Legenden und Labels können helfen.

## Tortendiagramm / Donut-Diagramm

Wir kommen nun zu einem sehr ähnlichen Diagrammtyp im Vergleich zu den Balkendiagrammen.



(from https://www.jku.at/studium/studienarten/bachelordiplom/ba-artificial-intelligence/)

**Eigenschaften eines Torten- bzw. Donut-Diagramms**:
* Kreissegmente stehen im Verhältnis zum Gesamtwert der Daten
* Darstellung relativer Anteile (oft in Prozent) gut geeignet (360° = 100%), absolute Werte seltener
* Kann als Torten-Diagramm (kompletter Kreis) oder als Donut-Diagramm (mit kreisförmiger Aussparung in der Mitte) dargestellt werden
* Farben zur Unterscheidung der Segmente werden häufig verwendet
* Zusätzliche Beschriftungen:
    * Titel
    * Werte in oder neben den Segmenten (oft in Prozent)
    * Legende zur Identifikation der Segmente
* Ideal für eine begrenzte Anzahl von Kategorien, da zu viele Segmente unübersichtlich wirken
* Donut-Diagramme bieten zusätzlichen Platz in der Mitte für Beschriftungen oder weiterführende Informationen



(von https://www.vienna.at/specials/nationalratswahl)

Was fällt uns hier auf? Warum haben wir hier nur einen Halbkreis?



(von https://www.vienna.at/sitzplan-im-nationalrat-steht-fest-wer-sitzt-neben-wem/6392113)

```python
# Wir können natürlich auch solche Diagramme mit Python erstellen

# Fiktive Daten
labels = ['Kategorie A', 'Kategorie B', 'Kategorie C', 'Kategorie D']
sizes = [25, 35, 20, 20]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

# Erstellen eines Tortendiagramms
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Tortendiagramm Beispiel")
plt.axis('equal')  # Sicherstellen, dass das Diagramm kreisförmig ist

# Anzeigen des Tortendiagramms
plt.show()
```

```python
# Erstellen eines Donut-Diagramms
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'width':0.4}) # Unterschied: wedgeprops={'width':0.4} !
plt.title("Donut-Diagramm Beispiel")
plt.axis('equal')  # Sicherstellen, dass das Diagramm kreisförmig ist

# Anzeigen des Donut-Diagramms
plt.show()
```

## Lineplot

Nun besprechen wir eine Möglichkeit, Daten zu visualisieren, welche mitunter auch geeignet ist, Funktionen zu plotten.



(von https://seaborn.pydata.org/generated/seaborn.lineplot.html)

**Eigenschaften eines Line Plots**:
* Darstellung von Datenpunkten als Linie, die durch Verbindung der Punkte die Veränderung einer Variablen über eine kontinuierliche Dimension (oft Zeit) zeigt
* Für nicht-kontinuierliche Daten ist ein Line-Plot sehr kontraproduktiv, da hier dann Werte dazwischen erwartet werden.
* Eignet sich besonders gut zur Visualisierung von Trends, Verläufen oder zeitlichen Entwicklungen
* Kann sowohl absolute als auch relative Werte darstellen
* Gut geeignet für den Vergleich mehrerer Datensätze, indem mehrere Linien in einem Diagramm verwendet werden
* Zusätzliche Beschriftungen:
    * Titel
    * Achsenbeschriftung (x-Achse für unabhängige Variable, y-Achse für abhängige Variable)
    * Legende zur Identifikation mehrerer Linien
    * Markierung der Datenpunkte kann hinzugefügt werden (z.B. Kreise, Punkte)
* Sehr intuitiv
* Unübersichtlich bei zu vielen Linien
* Empfindlich gegenüber Ausreißern



(from Marc Streit, VA WS2022; JKU)

```python
# Nun zeigen wir, wie wir in Python einen Lineplot erstellen können

data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Product A': [10, 20, 30, 25, 40],
    'Product B': [15, 25, 20, 35, 45],
    'Product C': [20, 30, 25, 30, 50]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))

plt.plot(df['Year'], df['Product A'], label='Product A', marker='o') # kreis
plt.plot(df['Year'], df['Product B'], label='Product B', marker='s') # square
plt.plot(df['Year'], df['Product C'], label='Product C', marker='d') # diamond

plt.title("Sales of Products Over Time", size=15)
plt.xlabel("Year", size=12)
plt.ylabel("Sales", size=12)

plt.legend()

plt.show()
```

```python
import numpy as np
```

```python
# Wir können natürlich so auch Funktionen plotten

def f(x):
    return 1/(1+np.exp(-x))

def f_prime(x):
    return f(x) * (1 - f(x))

x = np.linspace(-20, 20, 1000) # 1000 Werte zwischen -10 und 10
y = f(x)
y_prime = f_prime(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 1/(1+exp(-x))', color='red', marker='') # hier ist das Marker-Attribut eher kontraproduktiv
plt.plot(x, y_prime, label="f'(x) = f(x) * (1 - f(x))", color='blue', marker='') # hier ist das Marker-Attribut eher kontraproduktiv
plt.title('Sigmoid Function', size=15)
plt.xlabel('x', size=12)
plt.ylabel('f(x)', size=12)
plt.grid()
plt.legend()
plt.show()
```

### Korrelationskoeffizient

* Misst den linearen Zusammenhang zwischen 2 Datenreihen
* verschiedene Arten, wir verwenden (Bravais-Pearson Correlation Coefficient)
* Hat einen Wert zwischen $-1$ und $1$.
    * -1 bedeutet, es herrscht perfekte negative Korrelation
    * +1 bedeutet, es herrscht perfekte positive Korrelation
    * 0 bedeutet, es herrscht keine lineare Korrelation (kann aber trotzdem quadratisch sein)



(from https://en.wikipedia.org/wiki/Pearson_correlation_coefficient#/media/File:Correlation_examples2.svg)

Berechnung für Dataset $X=((x_1,y_1), (x_2,y_2), \ldots, (x_n,y_n))$:

\begin{align*}
    r_{xy} &= \frac{\textrm{Cov}(x,y)}{\sigma(x) \cdot \sigma(y)}\\
    &= \frac{\frac{1}{n}\sum_{i=1}^{n}(x_i-\overline{x})\cdot (y_i-\overline{y})}{\sqrt{\textrm{Var}(x)}\cdot \sqrt{\textrm{Var}(y)}}\\
    &= \frac{\frac{1}{n}\sum_{i=1}^{n}(x_i-\overline{x})\cdot (y_i-\overline{y})}{\sqrt{\textrm{Var}(x)\cdot \textrm{Var}(y)}} \\
    &= \frac{\frac{1}{n}\sum_{i=1}^{n}(x_i-\overline{x})\cdot (y_i-\overline{y})}{\sqrt{(\frac{1}{n}\sum_{i=1}^{n}(x_i-\overline{x})^2)\cdot (\frac{1}{n}\sum_{i=1}^{n}(y_i-\overline{y})^2)}}\\
    &= \frac{\sum_{i=1}^{n}(x_i-\overline{x})\cdot (y_i-\overline{y})}{\sqrt{(\sum_{i=1}^{n}(x_i-\overline{x})^2)\cdot (\sum_{i=1}^{n}(y_i-\overline{y})^2)}}
\end{align*}

Vergleich mit der bisher bekannten Methode (ergänzt um Korrelationskoeffizient)

```python
# regression including plot and its function without the confidence interval
def linear_regression(data: pd.DataFrame, x: str, y: str) -> None:
    """
    Perform linear regression on data. Makes a legend cell which contains the function of the regression.
    """
    # Create linear regression object
    regr = lm.LinearRegression()

    # Train the model using the training sets
    regr.fit(data[x].values.reshape(-1, 1), data[y])

    # Make predictions using the testing set
    y_pred = regr.predict(data[x].values.reshape(-1, 1))

    # Calculate Pearson correlation coefficient
    corr_coeff, _ = pearsonr(data[x], data[y])

    # Plot outputs
    plt.scatter(data[x], data[y])
    plt.plot(data[x], y_pred, color='blue', linewidth=3)

    # Add legend
    plt.legend([
        f"y = {regr.coef_[0]:.2f}x + {regr.intercept_:.2f}\n"
        f"r = {corr_coeff:.2f}"
    ])

    # add axis labels
    plt.xlabel(x)
    plt.ylabel(y)

    plt.show()
```

```python
linear_regression(data, 'Length', 'Weight')
```

## Radar Diagramm

Eine weitere Möglichkeit, mehrere Features pro Label/Kategorie darzustellen, ist das sogenannte Radar-Plot.

**Eigenschaften eines Radar-Diagramms**:
* Darstellung mehrerer Variablen in einem Netzdiagramm mit radialen Achsen
* Jeder Punkt auf den Achsen repräsentiert einen Wert, der durch Linien miteinander verbunden wird
* Gut geeignet, um mehrere Kategorien oder Objekte visuell miteinander zu vergleichen
* Normalerweise für relative Werte verwendet (z.B. Prozentsätze oder normierte Werte)
* Jeder Datensatz wird als eine Linie im Diagramm dargestellt, oft durch Farben unterschieden
* Weitere Beschriftungen:
    * Titel
    * Beschriftung der Achsen (jedes Segment beschreibt eine Kategorie)
    * Legende zur Identifikation der verschiedenen Datensätze
* Bei zu vielen Kategorien kann das Diagramm unübersichtlich werden
* Erneut eine Categorical Colormap zu verwenden
* Kann verzerrend sein, weil mehr Farbe (Kantenlänge) wenn weiter außen.

Nun benötigen wir das `plotly` Package.

Dieses lässt sich durch `conda install -c conda-forge plotly` installieren.

```python

data = {
    'r':[1, 5, 2, 2, 3],
    'theta':['processing cost','mechanical properties','chemical stability',
           'thermal stability', 'device integration']
}

df = pd.DataFrame(data)

fig = px.line_polar(df, r='r', theta='theta', line_close=True)
fig.show()
```

## Parallele Achsen Diagramm

Ein sehr ähnliches Konzept wie die Radar Diagramme sind die sogenannten Parallel-Coordinates Diagramme.



(from Marc Streit, VA WS22; JKU)

**Eigenschaften von Parallel Coordinates Plots**:
* Jede Variable wird als senkrechte Achse dargestellt, Werte werden durch Linien zwischen den Achsen verbunden
* Gut geeignet, um mehrere Variablen über verschiedene Datensätze hinweg zu vergleichen
* Jede Linie repräsentiert einen Datensatz und verbindet die Werte auf den entsprechenden Achsen
* Kann sowohl für absolute als auch relative Werte verwendet werden (Normalisierung nicht zwingend notwendig, weil jede Achse eine eigene Skalierung hat)
* Farbige Linien helfen, unterschiedliche Datensätze voneinander zu unterscheiden
* Weitere Beschriftungen:
    * Titel
    * Achsenbeschriftung (jede vertikale Achse steht für eine Variable)
    * Legende zur Identifikation verschiedener Datensätze
* Kann sowohl für große als auch kleine Datensätze verwendet werden, aber bei vielen Variablen oder Datensätzen kann es komplex werden (Overplotting)



(from Marc Streit, VA WS22; JKU)

```python
import plotly.express as px
import pandas as pd

# Beispiel-Datensatz erstellen
data = {
    'Feature 1': [0.1, 0.4, 0.5, 0.9],
    'Feature 2': [1.1, 3.4, 2.2, 1.5],
    'Feature 3': [2.0, 1.6, 3.3, 4.5],
    'Feature 4': [4.1, 2.5, 3.9, 2.2],
    'Label': ['A', 'B', 'C', 'D']
}
df = pd.DataFrame(data)

# Parallel Coordinates Plot erstellen
fig = px.parallel_coordinates(df,
                              color='Feature 1',  # Farbcodierung nach 'Feature 1'
                              labels={"Feature 1": "F1", "Feature 2": "F2", "Feature 3": "F3", "Feature 4": "F4"},
                              dimensions=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'],
                              color_continuous_scale=px.colors.diverging.Tealrose,
                              color_continuous_midpoint=2)

# Visualisierung anzeigen
fig.show()
```

## Heatmap

Auch bei dieser Art von Visualisierung ist es möglich, mehrere Labels mit mehreren Kategorien und deren Werte zu plotten. 



(von https://seaborn.pydata.org/generated/seaborn.heatmap.html)

**Eigenschaften einer Heatmap**:
* Darstellung von Daten in einem rechteckigen Gitter, bei dem die Zellen durch Farben die Werte repräsentieren
* Sequentielle Colormap
* Ideal zur Visualisierung von Datenmatrizen, um Muster, Korrelationen oder Anomalien zu erkennen
* Besonders gut geeignet für große Datenmengen, die schwer in herkömmlichen Diagrammen darstellbar sind
* Kann sowohl für absolute als auch für relative Werte verwendet werden
* Weitere Beschriftungen:
    * Titel
    * Achsenbeschriftung (x- und y-Achsen für Kategorien oder Zeiträume)
    * Farbskala oder Legende zur Interpretation der Farbwerte
* Oft in Kombination mit zusätzlichen Annotationen wie Zahlen in den Zellen für präzise Werte
* Schnelle Mustererkennung möglich
* Gut für Korrelationsanalysen zBsp.: Korrelationsmatrix
* Stark abhängig von Colormap

```python
glue = sns.load_dataset("glue").pivot(index="Model", columns="Task", values="Score")
```

```python
glue
```

```python
sns.heatmap(glue, annot=True, fmt=".1f")
```

```python
sns.heatmap(glue, annot=True, fmt=".1f", cmap='viridis')
```

```python
sns.heatmap(glue, annot=glue.rank(axis="columns")) # Auch Rank möglich!
```
